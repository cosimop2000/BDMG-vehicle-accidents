```python
# EDA
# Looking for cases where Humidity is zero and Precipitation is null (i.e., precipitation should be set to zero)
humidity_precip_filter = (accident_data['Humidity(%)'] == 0) & accident_data['Precipitation(in)'].isna()
result = accident_data[humidity_precip_filter][['Humidity(%)', 'Precipitation(in)']]
print(result)
```

```python
#looking for cases where wind speed is zero and wind direction is null (i.e., no wind to have a wind direction)
humidity_precip_filter = (accident_data['Wind_Speed(mph)'] == 0) & accident_data['Wind_Direction'].isna()
result = accident_data[humidity_precip_filter][['Wind_Speed(mph)', 'Wind_Direction']]
print(result)
```

```python
#looking to see if wind_speed is zero, is wind_chill null
#for these records, null wind_chill entries should be replaced with temperature
humidity_precip_filter = (accident_data['Wind_Speed(mph)'] == 0) & accident_data['Wind_Chill(F)'].isna()
result = accident_data[humidity_precip_filter][['Wind_Speed(mph)', 'Wind_Chill(F)']]
print(result)
```

```python
# EDA
# Looking to see if wind_speed is zero if there is a wind chill
wind_speed_chill_filter = (accident_data['Wind_Speed(mph)'] == 0) & accident_data['Wind_Chill(F)'].notna()
result = accident_data[wind_speed_chill_filter][['Wind_Speed(mph)', 'Wind_Chill(F)']]
print(result)
```

```python
# EDA
# Looking for null values in both wind speed and wind chill
null_values_filter = accident_data['Wind_Speed(mph)'].isna() & accident_data['Wind_Chill(F)'].isna()
result = accident_data[null_values_filter][['Wind_Speed(mph)', 'Wind_Chill(F)']]
print(result)
```
